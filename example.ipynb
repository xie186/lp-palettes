```python
from lets_plot.bistro.importer import get_data
mpg = get_data('mpg')
mpg.head()
```

## Applying discrete color scales (`scale_color_*`)

```python
(ggplot(mpg, aes(x='displ', y='hwy', color='class'))
 + geom_point(size=5)
 + scale_color_npg())
```

```python
(ggplot(mpg, aes(x='displ', y='hwy', color='class'))
 + geom_point(size=5)
 + scale_color_simpsons())
```

## Applying discrete fill scales (`scale_fill_*`)

```python
(ggplot(mpg, aes(x='class', fill='class'))
 + geom_bar()
 + scale_fill_lancet())
```

## Gallery of all palettes

```python
from lp_palettes.palettes import lp_palettes_db

for pal in lp_palettes_db.keys():
    for p in lp_palettes_db[pal].keys():
        scale_name = f"scale_color_{pal}"
        if pal == 'material':
            scale = globals()[scale_name](palette=p)
        else:
            scale = globals()[scale_name](palette=p) if p != 'default' else globals()[scale_name]()
        
        plot = (ggplot(mpg, aes(x='displ', y='hwy', color='class'))
                + geom_point(size=4)
                + scale
                + ggtitle(f'{pal}: {p}'))
        
        plot.show()
```
